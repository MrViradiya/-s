# **Task-2 : Image to Pencil Sketch with Python**

## **Importing libraries**

```python
# import necessary library for this project

import cv2 # opencv = open source computer vision
import matplotlib.pyplot as plt
%matplotlib inline
```

## **Read image**

```python
# read the image from the folder

read_img = cv2.imread("/content/khali.jpg")
```

## **Show image**

```python
# show the original image

show_img = plt.imshow(read_img)
plt.axis(False)
cv2.waitKey(0)  # display the window infinitely until any keypress
cv2.destroyAllWindows()
```

## **Convert image from BGR to RGB**

```python
# convert BGR image to RGB using cvtColor function

rgb_img = cv2.cvtColor(read_img,cv2.COLOR_BGR2RGB)

plt.imshow(rgb_img)
plt.axis(False)
plt.show()
```

## **Grey image**

```python
# create new image by converting original image to grey image

grey_img = cv2.cvtColor(rgb_img,cv2.COLOR_BGR2GRAY)


show_img = plt.imshow(grey_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## **Inverted grey image**

```python
# invert the grey image (also known as getting the image negative)

inverted_grey_img = 255-grey_img
# inverted_grey_img = cv2.bitwise_not(grey_img)

show_img = plt.imshow(inverted_grey_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## **Blur image**

```python
# blur the inverted grey image using gaussian function (also known as gaussian smoothing)

blur_img = cv2.GaussianBlur(inverted_grey_img, (21,21), 0)  # kernel size should be pair of odd numbers

show_img = plt.imshow(blur_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## **Inverted blur image**

```python
# invert the blur image

inverted_blur_img = 255 - blur_img

show_img = plt.imshow(inverted_blur_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## **Pencil sketch image**

```python
# create pencil sketch of inverted blur image

pencil_sketch_img = cv2.divide(grey_img,inverted_blur_img,scale=256)

show_img = plt.imshow(pencil_sketch_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## **Save pencil sketch image**

```python
cv2.imwrite('rock.jpg',pencil_sketch_img)
```

## **Original and Pencil sketch image**

```python
# show both original image and pencil sketch iamge

rgb_sketch = cv2.cvtColor(pencil_sketch_img,cv2.COLOR_BGR2RGB)

plt.subplot(1,2,1)
plt.title("Original image")
plt.imshow(rgb_img)
plt.axis(False)

plt.subplot(1,2,2)
plt.title("Pencil Sketch image")
plt.imshow(rgb_sketch)
plt.axis(False)

plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
